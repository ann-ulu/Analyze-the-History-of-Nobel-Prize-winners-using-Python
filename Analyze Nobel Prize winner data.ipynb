# Analyze Nobel Prize winner data and identify patterns by answering the following questions:

# What is the most commonly awarded gender?

```python
gender_count = nobel_data['sex'].value_counts()
gender_count

```

```python
sns.catplot(data=nobel_data, x="sex", kind="count")
plt.title("Award Distribution by Gender")
plt.show()


top_gender = nobel_data["sex"].value_counts().index[0]
top_gender_count = nobel_data["sex"].value_counts().iloc[0]
print('The top gender is', top_gender, "with a count of", top_gender_count)
```

```python
gender_count = nobel_data['sex'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(gender_count, labels=gender_count.index, autopct='%1.1f%%')
plt.title('Percent Distribution of Genders')
plt.show()
```

# What is the most commonly awarded birth country?


```python
most_awarded_birthcountry = nobel_data['birth_country'].value_counts()
most_awarded_birthcountry
```

```python
most_awarded_birthcountry = nobel_data['birth_country'].value_counts()
top_ten_countries = most_awarded_birthcountry[most_awarded_birthcountry >= 10]
df_top_ten_countries = top_ten_countries.reset_index()
df_top_ten_countries 
```

```python

df_top_ten_countries.columns = ['birth_country', 'count']

sns.catplot(x='birth_country', y='count', data=df_top_ten_countries, kind='bar', aspect = 2)
plt.xticks(rotation = 90)
plt.title('Award Distribution by Birth Country')
plt.show()

top_country = df_top_ten_countries['count'].iloc[0]
top_country_name = df_top_ten_countries['birth_country'].value_counts().index[0]
print("top_country is", top_country_name, "with", top_country, "awards" )
```

# Which decade had the highest ratio of US-born Nobel Prize winners to total winners in all categories?

```python
nobel_data['usa_born_winner'] = nobel_data['birth_country'] == 'United States of America'
nobel_data['decade'] = (nobel_data['year'] // 10) * 10

nobel_data
```

```python
usa_born_winners_per_decade_prob = nobel_data.groupby("decade", as_index=False)["usa_born_winner"].mean()
usa_born_winners_per_decade_prob

```

```python

sns.relplot(data=usa_born_winners_per_decade_prob, x="decade", y="usa_born_winner", kind="line", marker = "o")
plt.xticks(rotation=90)
plt.ylabel("US-born Nobel Prize Winners Ratio")
plt.title('Ratio of US brorn nobel prize winners by decade')
plt.show()


```

# Which decade had the highest proportion of female laureates?

```python
nobel_data['decade']
nobel_data['female winner per decade'] = nobel_data['sex'] == "Female"
female_winner_per_decade = nobel_data.groupby("decade", as_index = False)["female winner per decade"].mean()

female_winner_per_decade
```

```python
sns.lineplot(x="decade", y="female winner per decade", data=female_winner_per_decade, marker="o")

for x, y in zip(female_winner_per_decade["decade"], female_winner_per_decade["female winner per decade"]):
    plt.text(x, y, f'{y:.3f}', ha='right', va='bottom')
plt.title("Proportion of Female Awardee by decade")
plt.show()

decade_with_the_most_award = female_winner_per_decade['decade'].iloc[12]
print("The decade with the highest female awardeee is", decade_with_the_most_award )
```

# Which Nobel Prize category combination had the highest number of female laureates?

```python
female_awardee = nobel_data[nobel_data['sex'] == 'Female']
female_category_counts = female_awardee['category'].value_counts().reset_index()

female_category_counts

```

```python
female_category_counts.columns = ["category", "count"]

sns.barplot(x="category", y="count", data=female_category_counts)
plt.title("category Distribution for Female")
plt.show()

top_female_category = female_category_counts['category'][0]
top_female_category_counts = female_category_counts['count'][0]
print(" The top female award category is", top_female_category, "with a total award of", top_female_category_counts )
```

# Who was the first woman to receive a Nobel Prize, and in what category?

```python
# Filter for female laureates
female_awardee = nobel_data[nobel_data['sex'] == 'Female']

# Sort by year in ascending order
sorted_female_awardee = female_awardee.sort_values(by='year', ascending=True)
first_female_awardee = sorted_female_awardee.iloc[0]
first_female_name = first_female_awardee['full_name']
first_female_category = first_female_awardee['category']
first_female_year = first_female_awardee['year']

print("First Female Name: " + first_female_name + 
      "\nCategory: " + first_female_category + 
      "\nYear: " + str(first_female_year))
```
